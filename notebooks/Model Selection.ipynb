# Результаты оптимизаций optuna

Импорт библиотек и загрузка optuna studies:

```python
import os
import joblib
import optuna
import matplotlib
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
studies = []
studies_path = '../models/optuna_studies'
for file in os.listdir(studies_path):
    studies.append(joblib.load(os.path.join(studies_path, file)))
```

```python
os.listdir(studies_path)
```

## Случайный лес

Лучший показатель ROC AUC:

```python
studies[2].best_value
```

Лучшие параметры:

```python
studies[2].best_params
```

```python
optuna.visualization.matplotlib.plot_optimization_history(studies[2])
```

```python
optuna.visualization.matplotlib.plot_contour(studies[2]);
```

```python
optuna.visualization.matplotlib.plot_param_importances(studies[2])
```
